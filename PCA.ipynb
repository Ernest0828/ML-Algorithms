### STEP 2

Next, perform eigendecomposition of the covariance matrix estimated from the given data array. Finally, project original data onto lower-dimensional space and reconstruct data.

Proceed as follows:

1. Subtract the mean from the data, so that it is centered around the origin.

2. Estimate the covariance matrix from the centred data.

3. Calculate eigenvectors and eigenvalues using numpy functions

4. Project centered data (1845 dimension) into a lower-dimension space (You need to choose a reasonable dimension). 

5. Reconstruct the blood vessel shape from the lower dimension data in step 4.


As a sanity check plot a blood vessel shape reconstructed from the eigenvectors on top of the original blood vessel shape. Explain how much data reduction you have achieved. Comment on your results.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#PART 1
centred_data = blood_data - np.mean(blood_data, axis=0)
#Shape: (1845,7)

#PART 2
cov_matrix = np.cov(centred_data, rowvar=False)
print('Covariance Matrix:\n',cov_matrix)
#Shape: (7,7)

#PART 3
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

eigenval_df = pd.DataFrame(eigenvalues) #Variance
eigenval_df.columns = ['Eigenvalues']
print(eigenval_df)

eigenvec_df = pd.DataFrame(eigenvectors) #Principal components
print("Eigenvectors")
print(eigenvec_df)
```

```python
#PART 4 
projection_matrix = (eigenvectors.T[:][:3]).T

#PART 5
projected_data = np.dot(centred_data, projection_matrix)
reconstructed_data = np.dot(projected_data, projection_matrix.T)
reconstructed_data += np.mean(blood_data, axis=0)

#PLOTTING
fig = plt.figure(figsize=(15, 10))
ax3 = fig.add_subplot(projection='3d')

original_sample_2 = blood_data[:, 0]
original_sample = original_sample_2.reshape(615,3)
ax3.scatter(original_sample[:,0], original_sample[:,1], original_sample[:,2], c='blue', marker='x', label='Original')

reconstructed_sample = reconstructed_data[:,0].reshape(615,3)
ax3.scatter(reconstructed_sample[:, 0], reconstructed_sample[:, 1], reconstructed_sample[:, 2], c='red', marker='o', label='Reconstructed')

ax3.set_xlabel("x-coordinate")
ax3.set_ylabel("y-coordinate")
ax3.set_zlabel("z-coordinate")
ax3.set_title("Sample 1 Blood Shape - Original vs Reconstructed")

plt.legend()
plt.show()
```

```python
#INVESTIGATE VARIANCE 
import seaborn as sns

variance_explained = []
for i in eigenvalues:
     variance_explained.append((i/sum(eigenvalues)))        
print('Percentage variance explained\n',variance_explained)

cumulative_variance_explained = np.cumsum(variance_explained)
print(cumulative_variance_explained)

sns.lineplot(x = [1,2,3,4,5,6,7], y=cumulative_variance_explained)
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
plt.title("Explained variance vs Number of components")
```

### STEP 3

Research PCA analysis using the *scikit-learn* library. Perform PCA analysis and show the reconstructed data of any blood vessel shape on top of the original blood vessel shape. There are variables in the PCA object that correspond to the eigenvalues used for choosing projection eigenvectors. Compare the eigenvalues  and eigenvectors you have computed in the previous question with the eigenvalues  and the eigenvectors computed by the *scikit-learn* library. Compare the reconstructed coordinates from both methods. Comment on your results.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

blood_data_2 = np.load("shape_array.npy")

pca = PCA(n_components=3)
principal_components = pca.fit_transform(blood_data_2)

principal_components = pca.components_
principal_component_df = pd.DataFrame(principal_components)
principal_component_df.columns = ['Feature 1', 'Feature 2','Feature 3','Feature 4','Feature 5','Feature 6','Feature 7']
print(principal_component_df)

print('Explained Variance', pca.explained_variance_)

projected_data_2 = pca.fit_transform(blood_data_2)
projected_df = pd.DataFrame(projected_data_2, columns=['PC1','PC2','PC3'])
print(projected_df)
#Shape (1845,3)
```

```python
#Restores the dataset to original number of dimensions
reconstructed_data_2 = pca.inverse_transform(projected_data_2)
#Shape (1845,7)

#Compare Sample 1
original_sample = blood_data_2[:, 0].reshape(615, 3)
reconstructed_sample_2 = reconstructed_data_2[:, 0].reshape(615, 3)

# Plotting the original and reconstructed shapes
fig = plt.figure(figsize=(15, 10))

ax = fig.add_subplot(projection='3d')

# Plot original data 
ax.scatter(original_sample[:, 0], original_sample[:, 1], original_sample[:, 2], c='blue', marker='x', label='Original')

# Plot reconstructed data 
ax.scatter(reconstructed_sample_2[:, 0], reconstructed_sample_2[:, 1], reconstructed_sample_2[:, 2], c='red', marker='o', label='Reconstructed')

ax.set_xlabel('x-coordinate')
ax.set_ylabel('y-coordinate')
ax.set_zlabel('z-coordinate')
ax.set_title('Original vs Reconstructed Blood Vessel Shape for Sample 1')
ax.legend()

plt.show()
```
